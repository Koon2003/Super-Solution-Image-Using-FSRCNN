```python
import torch
import torchvision.transforms as T
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt
import torch.nn as nn
import torch.nn.functional as F
import math
from ignite.engine import Engine
from ignite.metrics import SSIM
```

```python
class FSRCNN(nn.Module):
    def __init__(self, scale_factor=2):
        super(FSRCNN, self).__init__()
        self.scale_factor = scale_factor
        self.conv1 = nn.Sequential(
                        nn.Conv2d(3, 56, kernel_size=5, padding=5//2),
                        nn.PReLU(56))
        self.conv2 = nn.Sequential(
                        nn.Conv2d(56, 12, kernel_size=1, padding=0),
                        nn.PReLU(12))
        self.conv3_1 = nn.Sequential(
                        nn.Conv2d(12, 12, kernel_size=3, padding=3//2),
                        nn.PReLU(12))
        self.conv3_2 = nn.Sequential(
                        nn.Conv2d(12, 12, kernel_size=3, padding=3//2),
                        nn.PReLU(12))
        self.conv3_3 = nn.Sequential(
                        nn.Conv2d(12, 12, kernel_size=3, padding=3//2),
                        nn.PReLU(12))
        self.conv3_4 = nn.Sequential(
                        nn.Conv2d(12, 12, kernel_size=3, padding=3//2),
                        nn.PReLU(12))
        self.conv4 = nn.Sequential(
                        nn.Conv2d(12, 56, kernel_size=1, padding=0),
                        nn.PReLU(56))
        self.conv5 = nn.ConvTranspose2d(56, 3, kernel_size=9, stride=scale_factor, padding=9//2, output_padding=scale_factor-1)
        self._initialize_weights()
    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3_1(x)
        x = self.conv3_2(x)
        x = self.conv3_3(x)
        x = self.conv3_4(x)
        x = self.conv4(x)
        x = self.conv5(x)
        return x
    def _initialize_weights(self) -> None:
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.normal_(m.weight.data, mean=0.0, std=math.sqrt(2 / (m.out_channels * m.weight.data[0][0].numel())))
                nn.init.zeros_(m.bias.data)

            elif isinstance(m, nn.ConvTranspose2d):
                nn.init.normal_(m.weight.data, mean=0.0, std=0.001)
                nn.init.zeros_(m.bias.data)
```

```python
def preprocess_image(image_path, scale_factor=2):
    # Mở ảnh từ file
    image = Image.open(image_path).convert("RGB")
    # Biến đổi: Resize ảnh đầu vào thành Low-Resolution (LR)
    transform_lr = T.Compose([
        T.Resize((image.size[1] // scale_factor, image.size[0] // scale_factor)),  # Resize theo scale_factor
        T.ToTensor()  # Chuyển đổi sang tensor
    ])
    transform_hr = T.ToTensor()  # Dùng để chuyển ảnh gốc thành tensor
    # Ảnh đầu vào (Low-Resolution)
    lr_image = transform_lr(image)
    # Ảnh gốc (High-Resolution)
    hr_image = transform_hr(image)
    # Thêm batch dimension: (B, C, H, W)
    lr_image = lr_image.unsqueeze(0)
    hr_image = hr_image.unsqueeze(0)
    return lr_image, hr_image, image
```

```python
def display_images(input_image, output_image, original_image):
    """Hiển thị ảnh đầu vào, ảnh đầu ra, và ảnh gốc."""
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    axes[0].imshow(input_image.permute(1, 2, 0).cpu().numpy())
    axes[0].set_title("Low-Resolution (Input)")
    axes[0].axis("off")
    
    axes[1].imshow(output_image.permute(1, 2, 0).cpu().detach().numpy())
    axes[1].set_title("Super-Resolution (Output)")
    axes[1].axis("off")
    
    axes[2].imshow(original_image)
    axes[2].set_title("High-Resolution (Original)")
    axes[2].axis("off")
    
    plt.show()
```

```python
# 1. Hàm process_function cho Engine
def process_function(engine, batch):
    y_pred, y = batch
    return y_pred, y

# 2. Tính toán PSNR
def calc_PSNR(lossMSE):
    return 20 * math.log10(1.0 / math.sqrt(lossMSE))

# 3. Tạo evaluator và gắn SSIM Metric
evaluator = Engine(process_function)
metric = SSIM(data_range=1.0)  # SSIM metric, giá trị pixel trong khoảng [0, 1]
metric.attach(evaluator, 'ssim')

# 4. Hàm tính SSIM
def calc_SSIM(preds, target):
    return evaluator.run([[preds, target]]).metrics['ssim']
```

```python
# Đường dẫn tới file model đã được huấn luyện
pretrained_model_path = "fsrcnn_trained_scale_2.pth"
scale_factor = 2   # Scale factor
# Khởi tạo mô hình FSRCNN
model = FSRCNN(scale_factor=scale_factor)  # Chọn scale_factor giống với lúc huấn luyện
# Load trọng số đã huấn luyện
model.load_state_dict(torch.load(pretrained_model_path))
# Chuyển mô hình sang chế độ đánh giá (evaluation)
model.eval()
```

```python
image_path = "test/0826.png"
# Tiền xử lý ảnh
lr_image, hr_image, original_image = preprocess_image(image_path, scale_factor=scale_factor)

# Dự đoán ảnh siêu phân giải
with torch.no_grad():
    sr_image = model(lr_image)  # Dự đoán
    # Đảm bảo kích thước khớp giữa SR và HR
    h1, w1 = sr_image.shape[2], sr_image.shape[3]  # Kích thước ảnh SR
    hr_image1 = hr_image[:, :, :h1, :w1]  # Cắt ảnh HR cho khớp kích thước

     # Tính PSNR và SSIM cho SR lần 1
    criterion = nn.MSELoss()
    mse_loss = criterion(sr_image, hr_image1)
    psnr_value = calc_PSNR(mse_loss.item())
    
    ssim_value = calc_SSIM(sr_image, hr_image1)
    
    print(f"PSNR lần giữa sr với hr: {psnr_value:.2f} dB")
    print(f"SSIM lần giữa sr với hr: {ssim_value:.4f}")

    # Chuyển ảnh SR lần 1 làm đầu vào cho lần 2
    sr_image1 = torch.clamp(sr_image, 0, 1)  # Đảm bảo giá trị pixel hợp lệ

display_images(lr_image[0], sr_image1[0], original_image)
```
